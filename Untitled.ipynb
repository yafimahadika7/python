```python
print("\nAnalisis nilai kosong (NaN) pada setiap kolom:")
missing_data = data_startup.isna().sum()
print(missing_data)
```

```python
numeric_columns = data_startup.select_dtypes(include=['float64', 'int64']).columns
data_startup[numeric_columns] = data_startup[numeric_columns].fillna(data_startup[numeric_columns].mean())
```

```python
print("\nKolom setelah mengisi nilai NaN dengan mean:")
missing_data_after_filling = data_startup.isna().sum()
print(missing_data_after_filling)
```
